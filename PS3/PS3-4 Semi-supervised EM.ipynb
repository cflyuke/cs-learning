### (c)

(i)首先求$\phi$，注意$\sum_{l=1}^k\phi_l = 1$，所以要用拉格朗日乘子法取计算。不难得到一下结果
$$\phi_j = \frac{\sum_{i=1}^m w_j^{(i)}+\alpha \sum_{i=1}^m1
\{\tilde{z}^{(i)} = j\}} {m + \alpha\tilde{m}} $$

(ii)对于$\mu_j$直接求偏导即可
$$ \mu_j = \frac{\sum_{i = 1}^m w_j^{(i)}x^{(i)} + \alpha \sum_{i=1}^m1\{z^{(i)} = j\}\tilde{x}^{(i)}}{ \sum_{i = 1}^m w_j^{(i)} + \alpha \sum_{i=1}^m1\{z^{(i)} = j\}}
$$

(iii)对与$\Sigma$进行求偏导时，要用到以下公式:
$$ \nabla_{A} tr(BA^{-1}) = -A^{-T}BA^{-1} $$
从而求偏导可得：
\begin{align*}
  \nabla_{\Sigma_j} l &= -\frac{1}{2} \Sigma_j^{-1}(\sum\limits_{i=1}^mw_j^{(i)} + \alpha \sum_{i=1}^{\tilde{m}} 1\{\tilde{z}^{(i)} = j\}) \\
& + \frac{1}{2}\Sigma_j^{-1}(\sum_{i=1}^mw_j{(i)}(x^{(i)}-\mu_j)(x^{(i)}-\mu_j)^T + \alpha\sum_{i=1}^{\tilde{m}}1\{\tilde{z}^{(i)} = j\}(\tilde{x}^{(i)}-\mu_j)(\tilde{x}^{(i)}-\mu_j)^T)\Sigma_j^{-1} = 0 \\
\Sigma_j &= \frac{\sum_{i=1}^mw_j{(i)}(x^{(i)}-\mu_j)(x^{(i)}-\mu_j)^T + \alpha\sum_{i=1}^{\tilde{m}}1\{\tilde{z}^{(i)} = j\}(\tilde{x}^{(i)}-\mu_j)(\tilde{x}^{(i)}-\mu_j)^T}{\sum\limits_{i=1}^mw_j^{(i)} + \alpha \sum_{i=1}^{\tilde{m}} 1\{\tilde{z}^{(i)} = j\}}
\end{align*}
