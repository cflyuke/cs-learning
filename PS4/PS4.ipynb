## PS4-6  Reinforcement Learning: The inverted pendulum

整个强化学习过程的大概分为以下几步
- 初始化阶段
  1. 设置随机种子（每次实验得到相同的随机序列，方便对他人的实验结果进行复现）
  2. 初始化模拟参数
  3. 初始化计数器（总共进行了多少次实验达到收敛，每次实验进行了多少步失败）
  4. 创建交互环境
  5. 初始化MDP数据，包括转移概率，奖励函数，value等信息
- 训练阶段
  1. 选择当前状态所能采取的最优动作
  2. 与环境交互，得到反馈信息
  3. 更新MDP模型（transition_counts和reward_counts）
  4. 失败处理，判断是否失败，更新MDP值，并分析是否已达到收敛
